```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager
from matplotlib.pyplot import MultipleLocator
```

```python
# prepare for calibri fonts
!gdown "1vS4zImMQE4DsrVWlsn52rkDaCM_vmJba" --output calibri.ttf
!gdown "1vqxCJ90X4h9-IjMO0MALq6XwXGiv9p5S" --output calibri-bold.ttf

font_manager.fontManager.addfont("calibri.ttf")
font_manager.fontManager.addfont("calibri-bold.ttf")

plt.rcParams['font.family'] = "calibri"
plt.rcParams['font.weight'] = "bold"
```

```python
x_data = [1, 2, 3, 4, 5, 6]
y_data = [38.73, 38.71, 38.70, 38.68, 38.65, 38.49]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, sharex=False, sharey=False, figsize=(12, 9))
ax.plot(x_data, y_data, color="#f39a1e", alpha=1., linewidth=6, marker="o", markersize=18, zorder=3)

# spines
for axis in ["top", "bottom", "left", "right"]:
    ax.spines[axis].set_visible(True)
    ax.spines[axis].set_linewidth(4.5)

# title
ax.set_title(label="PSNR-Pruning ratio plot", pad=30, loc="center", size=32, weight="bold")

# x-axis
ax.set_xlabel(xlabel="Pruning ratio", labelpad=20, size=28, weight="bold")
ax.set_xlim(left=0.3, right=6.7)  # or ax.margins(x=0.2)
ax.set_xscale("linear")  # "linear", "log", "symlog"...
ax.tick_params(axis="x", which="major", reset=False, direction="in", length=8, width=4.5, pad=15)
ax.set_xticklabels(ax.get_xticks(), rotation=0, size=24, weight='bold')
ax.grid(visible=True, axis="x", which="major", linewidth=1, linestyle="-", zorder=0)
ax.xaxis.set_major_locator(MultipleLocator(1))

# y-axis
ax.set_ylabel(ylabel="PSNR (dB)", labelpad=20, size=28, weight="bold")
ax.set_ylim(bottom=37., top=39.5)  # or ax.margins(y=0.2)
ax.set_yscale("linear")  # "linear", "log", "symlog"...
ax.tick_params(axis="y", which="major", reset=False, direction="in", length=8, width=4.5, pad=15)
ax.set_yticklabels(ax.get_yticks(), rotation=0, size=24, weight='bold')
ax.grid(visible=True, axis="y", which="major", linewidth=1, linestyle="-", zorder=0)
# ax.yaxis.set_major_locator(MultipleLocator(0.5))


# set aspect ratio (Y : X)
ratio = 3 / 4
x_left, x_right = ax.get_xlim()
y_low, y_high = ax.get_ylim()
ax.set_aspect(abs((x_right-x_left)/(y_low-y_high)) * ratio)

fig.tight_layout()
plt.savefig("plot.png")
```
